```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./데이터자료/5. 이자보상배율 완료.csv",encoding='cp949')
df
```

```python
df.reset_index(inplace=True)
df_label = df[['index','회사명', '거래소코드', '회계년도', '상장일', '상장폐지일', '소속코드']]
```

```python
df_label= df_label[df_label['소속코드']==4]
df_label= df_label.groupby('회사명')['index'].apply(list)
```

```python
df_label = pd.DataFrame(df_label)
df_label
```

```python
for j in df_label.index:
    for i in df.index:
        if df_label["index"][j][-1] == i:
            df.loc[i,'label'] = 1
df['label'] = df["label"].fillna(0)
```

# label 설정된 부실기업은 부실 해당년도 제외하고 삭제

```python
defalut = df[~df["상장폐지일"].isnull()]
defalut
```

```python
drop_index=defalut[defalut['label']==0].index
drop_index
```

```python
df.drop(drop_index, inplace=True)
```

```python
df.drop('index',axis=1,inplace=True)
```
