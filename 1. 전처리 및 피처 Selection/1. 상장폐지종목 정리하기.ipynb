# 거래소코드 중복 데이터 확인해서 지우기

```python
b = df[["회사명","거래소코드"]]
d = b.groupby("회사명")["거래소코드"].unique()
d = pd.DataFrame(d)
d
```

```python
for i in d.index:
    if len(d.loc[i,"거래소코드"])== 2:
        print(i)
```
