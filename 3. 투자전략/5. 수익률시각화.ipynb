```python
def sharpe(data):
# 전략누적수익률의 표준편차를 계산한다
    Std = (data["포트폴리오 일일수익률"]).std( )

    # 전략의 샤프비율을 구하고 출력한다
    Sharpe = ( data["포트폴리오 일일수익률"] - data["kosdaq"] )/Std
    Sharpe =Sharpe.mean()
    # print(((100+data["누적수익률(%)"].min())-(100+data["누적수익률(%)"].max()))/(data["누적수익률(%)"].max()+100))
    print(Std)
    print( 'Sharpe ratio: %.2f' % Sharpe )
    print(Sharpe *np.sqrt(252))
```
