```python
#pykrx 데이터 끌고오기
import pandas as pd
import numpy as np
import FinanceDataReader as fdr
```

```python
data = pd.read_csv("./데이터/5. label 지정 완료.csv", encoding='cp949')
data
```

```python
# 거래소 코드 6개로 맞춰주기
for i in data.index:
    if len(str(data.loc[i,"거래소코드"])) == 3:
        data.loc[i,"거래소코드"]='000'+str(data.loc[i,"거래소코드"])
    elif len(str(data.loc[i,"거래소코드"])) == 4:
        data.loc[i,"거래소코드"]='00'+str(data.loc[i,"거래소코드"])
    elif len(str(data.loc[i,"거래소코드"])) == 5:
        data.loc[i,"거래소코드"]='0'+str(data.loc[i,"거래소코드"])
    elif len(str(data.loc[i,"거래소코드"])) == 2:
        data.loc[i,"거래소코드"]='0000'+str(data.loc[i,"거래소코드"])
```

```python
#거래소 코드 개별로 추출
a =data["거래소코드"].unique()
a = pd.DataFrame(a)
a.columns=["거래소코드"]
len(a)
```

```python
# 코스닥 개별기업들의 일자별 주가 추출
b=[]
for i in a.index:
    df = fdr.DataReader(str(a.loc[i,"거래소코드"]),"2011-01-01", "2021-12-31")
    b.append(df)
```

```python
# 각 인덱스에 회사명이랑 거래소코드 넣어주기
for i in a.index:
    b[i]["거래소코드"]=a.loc[i,"거래소코드"]
    b[i]["회사명"] =a.loc[i,"회사명"]
```

```python
# 기업별 데이터프레임 합하기
data_all = pd.concat(b)
```

```python
data_all=data_all.reset_index()
data_all
```

```python
data_all.to_csv("./데이터/일일주식데이터.csv", encoding='cp949')
```
