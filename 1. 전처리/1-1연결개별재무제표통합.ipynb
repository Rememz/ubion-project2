## 데이터 불러오기
- TS에서 금융업 제외 및 결산기 12월인 데이터만 추출

```python
df = pd.read_csv('./연결재무제표.csv', encoding='CP949')
df.head()
```

### 연결재무제표의 결측치를 개별재무제표의 값으로 바꾸기

```python
#컬럼명 일치를 위한 컬럼명 변경 
df.columns = df.columns.str.replace('연결','')
df.columns 
```

```python
df_na=pd.read_csv('./개별재무제표.csv', encoding='CP949')
df_na.head()
```

```python
# 컬럼명 리스트로 변경
list = df.columns.tolist()
```

```python
# 빈공간을 개별재무제표값으로 넣기
for i in list:
    df[i] = df[i].fillna(df_na[i])
```

```python
df["회계년도"]=pd.to_datetime(df['회계년도'])
df
```

## 부도기업중 상폐사유가 맞지 않은 데이터 삭제
- 피흡수합병, 유가시장상장등 사유의 상폐기업은 제외
- csv파일에서 대상 기업들에게 삭제여부 컬럼에 삭제를 배정

```python
df_del = pd.read_csv('./부도기업사유.csv', encoding='CP949')
df_del = df_del.set_index(['Unnamed: 0'])
df_del["회계년도"]=pd.to_datetime(df_del['회계년도'], format='%y-%b')
df_del
```

```python
drop_del = df_del[df_del['삭제여부']=='삭제'].index
df = df.drop(drop_del)
df
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```
