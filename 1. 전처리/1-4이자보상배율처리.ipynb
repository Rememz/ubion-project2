```python
import pandas as pd
import numpy as np
```

```python
# 이자보상배율 
rate = pd.read_csv('./데이터자료/이자보상배율구하기.csv', encoding='CP949')
rate['회계년도'] = pd.to_datetime(rate['회계년도'], format='%b-%y')
df = pd.read_csv('./데이터자료/label(완).csv', encoding='cp949',index_col=0)
df['회계년도'] = pd.to_datetime(df['회계년도'])
#컬럼명 일치를 위한 컬럼명 변경 
rate.columns = rate.columns.str.replace('   ','')
rate.columns 
```

```python
rate['이자보상배율'] = rate['[제조]영업이익(손실)(연결)']/rate['[제조]이자비용(연결)']
```

```python
inf_rate = rate[rate['[제조]이자비용(연결)']==0]
a12= rate[rate['[제조]이자비용(연결)']==0].index
not_inf_rate = rate.drop(a12)
```

```python
rate[rate["회사명"]=="(주)팍스넷"]
```

```python
#이자보상배율이상치 확인
df_n = df[df['[제조]이자보상배율(이자비용)']==999999999]
df_ni = df[df['[제조]이자보상배율(이자비용)']==999999999].index
df_nn = df.drop(df_ni)
```

```python
df_n[df_n["회사명"]=="(주)팍스넷"]
```

```python
dfa = df_nn.groupby(['회사명']).mean()
dfa = dfa[["거래소코드",'[제조]이자보상배율(이자비용)']]
```

```python
inf_dfa = pd.merge(inf_rate,dfa,on='거래소코드',how='inner')
inf_dfa.rename(columns={'[제조]이자보상배율(이자비용)':'수정[제조]이자보상배율(이자비용)'},inplace=True)
inf_dfa.drop(['[제조]영업이익(손실)(연결)','[제조]이자비용(연결)','이자보상배율'],axis=1,inplace=True)
inf_dfa
```

```python
acc = inf_dfa.index
acc
```

```python
dfa_ija = pd.merge(inf_rate,inf_dfa, on=['회사명','거래소코드','회계년도'], how='inner')
dfa_ija.drop(['[제조]영업이익(손실)(연결)', '[제조]이자비용(연결)', '이자보상배율'], axis=1, inplace=True)
axd = inf_rate['회사명'].tolist()
aqd = dfa_ija['회사명'].tolist()
```

```python
list = set(axd) - set(aqd)
list
```

```python
inf_d = inf_rate[inf_rate['회사명'].isin(list)]
dc = df_nn['[제조]이자보상배율(이자비용)'].max()
dm = df_nn['[제조]이자보상배율(이자비용)'].min()
```

```python
am = inf_d[inf_d['[제조]영업이익(손실)(연결)'] < 0]
am['수정[제조]이자보상배율(이자비용)'] = -4493653.05
ad = inf_d[inf_d['[제조]영업이익(손실)(연결)'] > 0]
ad['수정[제조]이자보상배율(이자비용)'] = 8432098.32
```

```python
inf_d = pd.concat([am,ad])
inf_d.drop(['[제조]영업이익(손실)(연결)', '[제조]이자비용(연결)', '이자보상배율'],axis=1, inplace=True)
inf_dfas = pd.concat([inf_dfa,inf_d])
inf_dfas

```

```python
pd.options.display.float_format = '{:.5f}'.format#무슨뜻?
```

```python
inf_dfas.rename(columns={'수정[제조]이자보상배율(이자비용)':'이자보상배율'},inplace=True)
ija_com = pd.concat([not_inf_rate,inf_dfas])
ija_com.sort_values(['회사명','회계년도'], ascending=True, inplace=True)
ija_com
```
