```python
data = pd.read_csv('./데이터자료/dataset.csv', encoding='CP949',index_col=0)
data2 = pd.read_csv("./데이터자료/데이터전처리자료(10년도포함).csv", encoding='cp949')

#컬럼명 일치를 위한 컬럼명 변경 
```

```python
#30 컬럼명 간단하게 변경
data2.columns =['회사명', '거래소코드', '회계년도', '매출액(수익)','매출액(수익)(연결)', '매출액증가율', '매출액증가율(연결)',
       '유형자산', '유형자산(연결)', '유형자산증가율','유형자산증가율(연결)', '재고자산', '재고자산(연결)',
       '재고자산증가율', '재고자산증가율(연결)', '비유동자산', '비유동자산(연결)', '비유동자산증가율(연결)', '유동자산',
       '유동자산(연결)', '유동자산증가율(연결)', '자본', '자본(연결)', '자기자본증가율(연결)', '총자본증가율', '자산', '자산(연결)', '총자본증가율(연결)',
       '당기순이익(총포괄손익)', '당기순이익(총포괄손익)(연결)', '순이익증가율', '순이익증가율(연결)', '비유동자산증가율', '유동자산증가율', '자기자본증가율']
```

```python
data3= data2[['회사명', '거래소코드', '회계년도',"매출액(수익)(연결)","유형자산(연결)","재고자산(연결)","비유동자산(연결)",
"유동자산(연결)","자본(연결)","자산(연결)","당기순이익(총포괄손익)(연결)"]]
```

```python
data3["매출액(수익)(연결)"] = data3["매출액(수익)(연결)"].fillna(data2['매출액(수익)'])
data3["유형자산(연결)"] = data3["유형자산(연결)"].fillna(data2['유형자산'])
data3["재고자산(연결)"] = data3["재고자산(연결)"].fillna(data2['재고자산'])
data3["비유동자산(연결)"] = data3["비유동자산(연결)"].fillna(data2['비유동자산'])
data3["유동자산(연결)"] = data3["유동자산(연결)"].fillna(data2['유동자산'])
data3["자본(연결)"] = data3["자본(연결)"].fillna(data2['자본'])
data3["자산(연결)"] = data3["자산(연결)"].fillna(data2['자산'])
data3["당기순이익(총포괄손익)(연결)"] = data3["당기순이익(총포괄손익)(연결)"].fillna(data2['당기순이익(총포괄손익)'])
```

```python
data3= data3.dropna()
data3 = data3.reset_index(drop='index')
```

```python
data3
```

```python
list = data3.iloc[:,3:11].columns
list
```

```python
# 값 계산하기 
for i in data3.index:
    for j in list:
        if i > 0:
            if data3.iloc[:,1][i] == data3.iloc[:,1][i-1]: 
                data3.loc[i,j+'증가율'] = (data3.loc[i,j]-data3.loc[i-1,j])/data3.loc[i-1,j]*100
            elif data3.iloc[:,1][i] != data3.iloc[:,1][i-1]: 
                data3.loc[i,j+'증가율'] = 0
        elif i == 0:
            data3.loc[i,j+'증가율'] = 0
```

```python
# 재고자산, 자본, 순이익에 대한 전처리
for i in data3.index:
    if i>0 :
        if data3.iloc[:,5][i] == 0:
            data3.loc[i,'재고자산(연결)증가율']=0
        elif data3.iloc[:,5][i-1] == 0:
            data3.loc[i,'재고자산(연결)증가율']=0
        if data3.iloc[:,8][i] <=0:
            data3.loc[i,"자본(연결)증가율"] = 0
        elif data3.iloc[:,8][i-1] <=0:
            data3.loc[i,"자본(연결)증가율"] = 0
        if data3.iloc[:,10][i] <=0:
            data3.loc[i,'당기순이익(총포괄손익)(연결)증가율'] = 0
        elif data3.iloc[:,10][i-1] <=0:
            data3.loc[i,'당기순이익(총포괄손익)(연결)증가율'] = 0
```

```python

data3=data3[~data3.isin([ np.inf, -np.inf]).any(1)]

data3
```

```python
# 0인 데이터들 삭제
data3_10 = data3[data3["회계년도"]=='2010/12'].index
data3 = data3.drop(data3_10)
data3_del = data3[data3["매출액(수익)(연결)증가율"]==0].index
data3 = data3.drop(data3_del)
```

```python
data3 = data3.reset_index(drop='index')
data4 = data3.iloc[:,[0,1,2,11,12,13,14,15,16,17,18]]
data4 = data4.round(2)
data4
```

```python
data4["회계년도"]=pd.to_datetime(data4['회계년도'])
data["회계년도"]=pd.to_datetime(data['회계년도'])
```

```python
data4.to_csv("증가율데이터.csv", encoding='cp949')
```

```python
df = pd.merge(data,data4,how='left', on=["회사명","거래소코드","회계년도"])
```

```python
df_del=df[df['매출액(수익)(연결)증가율'].isnull()].index
df = df.drop(df_del)
df = df.reset_index(drop='index')
df
```

```python
df["[제조]매출액증가율"] = df['매출액(수익)(연결)증가율']
df["[제조]유형자산증가율"] = df['유형자산(연결)증가율'].fillna(0)
df["[제조]재고자산증가율"] = df['재고자산(연결)증가율']
df["[제조]비유동자산증가율"] = df['비유동자산(연결)증가율']
df["[제조]유동자산증가율"] = df['유동자산(연결)증가율']
df["[제조]자기자본증가율"] = df['자본(연결)증가율']
df["[제조]총자본증가율"] = df['자산(연결)증가율']
df["[제조]매출액순이익률"] = df['당기순이익(총포괄손익)(연결)증가율']
```
