## 연결데이터 기준으로 증가율 데이터 생성하기
### 재무데이터 마다 특성이 다르기 때문에 컬럼 하나하나로 수정할것!!!
### 회사명이 다를때는 회사의 첫 데이터에 NaN값 넣기

```python
# 1. 매출액 데이터 : 0이 없기 때문에 일반적으로 계산 
df_data[df_data["매출액(수익)(*)(IFRS연결)(천원)"]==0]
# df_data["매출액(수익)(*)(IFRS연결)(천원)"].unique()
```

```python
# iloc 모르면 물어볼 것
# 데이터에 0 이 있기 때문에 데이터데 계산데이터에 0이 들어있으면 증가율 데이터는 무조건 0처리
for i in df_data.index:
    if i > 0:
        if df_data.loc[i,"회사명"] == df_data.loc[i-1,"회사명"]:
            if df_data.iloc[:,4][i] == 0 or df_data.iloc[:,4][i-1] == 0:
                df_data.loc[i,"매출액증가율"] = 0
            else:
                df_data.loc[i,"매출액증가율"] = (df_data.iloc[:,4][i]-df_data.iloc[:,4][i-1])/df_data.iloc[:,4][i-1]
        elif df_data.loc[i,"회사명"] != df_data.loc[i-1,"회사명"]:
            df_data.loc[i,"매출액증가율"] = np.NAN
```

```python
# 6. 자기자본
# 자기자본에는 0 또는 마이너스 값인 데이터 존재 이 -값들이 들어가는 계산에 대해서는 모두 0으로 처리
print(df_data["자본(*)(IFRS연결)(천원)"].unique())
df_data[df_data["자본(*)(IFRS연결)(천원)"]<=0]
```

```python
# 자기자본에는 0 또는 마이너스 값인 데이터 존재 이 -값들이 들어가는 계산에 대해서는 모두 0으로 처리
for i in df_data.index:
    if i > 0:
        if df_data.loc[i,"회사명"] == df_data.loc[i-1,"회사명"]:
            if df_data.iloc[:,14][i] <= 0 or df_data.iloc[:,14][i-1] <= 0:
                df_data.loc[i,"자기자본증가율"] = 0
            else:
                df_data.loc[i,"자기자본증가율"] = (df_data.iloc[:,14][i]-df_data.iloc[:,14][i-1])/df_data.iloc[:,14][i-1]
        elif df_data.loc[i,"회사명"] != df_data.loc[i-1,"회사명"]:
            df_data.loc[i,"자기자본증가율"] = np.NAN
```

```python
# 7. 총자본
print(df_data["자산(*)(IFRS연결)(천원)"].unique())
df_data[df_data["자산(*)(IFRS연결)(천원)"]<0]
```

```python
# 총자본값에는 0이 존재하지 않아서 일반으로 계산
for i in df_data.index:
    if i > 0:
        if df_data.loc[i,"회사명"] == df_data.loc[i-1,"회사명"]:
                df_data.loc[i,"총자본증가율"] = (df_data.iloc[:,16][i]-df_data.iloc[:,16][i-1])/df_data.iloc[:,16][i-1]
        elif df_data.loc[i,"회사명"] != df_data.loc[i-1,"회사명"]:
            df_data.loc[i,"총자본증가율"] = np.NAN
```

```python
# 8. 순이익
# 순이익의 -값이 상당히 많음
print(df_data["당기순이익(총포괄손익)(IFRS연결)(천원)"].unique())
df_data[df_data["당기순이익(총포괄손익)(IFRS연결)(천원)"]<=0]
```

```python
# 우리조에서는 당기순이익의 -에서 값이 줄어드는 거는 부도기업의 정의에서 상황이 나아졌다고 판단하기 어렵다고 판단해서 순이익이 -이면 0으로 판단
for i in df_data.index:
    if i > 0:
        if df_data.loc[i,"회사명"] == df_data.loc[i-1,"회사명"]:
            if df_data.iloc[:,18][i] <= 0 or df_data.iloc[:,18][i-1] <= 0:
                df_data.loc[i,"순이익증가율"] = 0
            else:
                df_data.loc[i,"순이익증가율"] = (df_data.iloc[:,18][i]-df_data.iloc[:,18][i-1])/df_data.iloc[:,18][i-1]
        elif df_data.loc[i,"회사명"] != df_data.loc[i-1,"회사명"]:
            df_data.loc[i,"순이익증가율"] = np.NAN
```
